```python

#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Mon Jul 15 14:50:39 2019

@author: aarthi
"""


import metquest
import networkx as nx
from collections import Counter


def find_metabolic_support_index(pname, graph_name_combined, graph_name_firstorg, graph_name_secondorg, seed_metabolite_fname):
    '''
    This function evaluates the metabolic support index which defines the
    Fraction of blocked or stuck reactions relieved by the presence of another
    organism.

    Parameters
    ----------

    pname - Pathname with all the datafiles
    graph_name_combined - Filename of the combined graphs
    graph_name_firstorg - Filename of the first organism graph
    graph_name_secondorg - Filename of the second organism graph
    seed_metabolite_fname - Filename of the seed metabolite list

    Returns
    -------
    None

    '''
    combined_graph = nx.read_gpickle(pname + graph_name_combined) #seed_mets.txt'
    first_org_graph = nx.read_gpickle(pname + graph_name_firstorg)
    second_org_graph = nx.read_gpickle(pname + graph_name_secondorg)
    with open(pname + seed_metabolite_fname) as f: # 'pichia_higher_benefit_seed_1.txt' original_seed_mets.txt'pichia_higher_benefit_seed.txt''original_seed_mets.txt''original_seed_mets.txt'
        seed_metabolites = f.read().splitlines()
    seed_metabolites = set(seed_metabolites)

    lb_m, status_dict, scope = metquest.forward_pass(combined_graph, seed_metabolites)
    lb_m_org_1, status_dict_org_1, scope_org_1 = metquest.forward_pass(first_org_graph, seed_metabolites)
    org_1_reactions = []
    org_1_metabolites = []
    node_attr = nx.get_node_attributes(first_org_graph, 'bipartite')
    for nodes in first_org_graph:
        if node_attr[nodes] == 1:
            org_1_reactions.append(nodes)
        else:
            metname = nodes.split(' ')[0]
            if metname not in org_1_metabolites and '_e' not in metname:
                org_1_metabolites.append(metname)

    lb_m_org_2, status_dict_org_2, scope_org_2 = metquest.forward_pass(second_org_graph, seed_metabolites)
    org_2_reactions = []
    org_2_metabolites = []
    node_attr = nx.get_node_attributes(second_org_graph, 'bipartite')
    for nodes in second_org_graph:
        if node_attr[nodes] == 1:
            org_2_reactions.append(nodes)
        else:
            metname = nodes.split(' ')[0]
            if metname not in org_2_metabolites and '_e' not in metname:
                org_2_metabolites.append(metname)
    org_1 = org_1_metabolites[0]
    org_2 = org_2_metabolites[0]
    reactions = []
    org_1_reactions_in_combined = []
    org_2_reactions_in_combined = []
    node_attr = nx.get_node_attributes(combined_graph, 'bipartite')
    for nodes in combined_graph:
        if node_attr[nodes] == 1:
            reactions.append(nodes)
        if org_1 in nodes:
            org_1_reactions_in_combined.append(nodes)
        if org_2 in nodes:
            org_2_reactions_in_combined.append(nodes)
    org_1_internal_reactions_in_combined = []
    dummy_1 = []
    for rxns in org_1_reactions_in_combined:
        if 'ER' not in rxns:
            org_1_internal_reactions_in_combined.append(rxns)
        else:
            dummy_1.append(rxns)
    org_2_internal_reactions_in_combined = []
    dummy_2 = []
    for rxns in org_2_reactions_in_combined:
        if 'ER' not in rxns:
            org_2_internal_reactions_in_combined.append(rxns)
        else:
            dummy_2.append(rxns)

    # Combined stuck nodes
    all_stuck_nodes = set(reactions) - set(status_dict)
    stuck_internal_rxns_in_org_1_combined = []
    stuck_internal_rxns_in_org_2_combined = []
    stuck_external_rxns_in_org_1_combined = set([])
    stuck_external_rxns_in_org_2_combined = set([])
    for nodes in all_stuck_nodes:
        if org_2 in nodes:
            if 'NCE' not in nodes and 'ER' not in nodes:
                if 'IR' in nodes or 'RR' in nodes or 'RevBR' in nodes:
                    stuck_internal_rxns_in_org_2_combined.append(nodes)
            else:
                stuck_external_rxns_in_org_2_combined.add(nodes)
        elif org_1 in nodes:
            if 'NCE' not in nodes and 'ER' not in nodes:
                if 'IR' in nodes or 'RR' in nodes or 'RevBR' in nodes:
                    stuck_internal_rxns_in_org_1_combined.append(nodes)
            else:
                stuck_external_rxns_in_org_1_combined.add(nodes)

    org_2_stuck_nodes = set(org_2_reactions) - set(status_dict_org_2)
    org_2_stuck_internal_single = []
    org_2_stuck_external = []
    for nodes in org_2_stuck_nodes:
        if 'IR' in nodes or 'RR' in nodes or 'RevBR' in nodes:
            if 'E' not in nodes:
                org_2_stuck_internal_single.append(nodes)
            else: #ERR nodes
                org_2_stuck_external.append(nodes)
        else:
            org_2_stuck_external.append(nodes)
    org_1_stuck_nodes_internal = set(org_1_reactions) - set(status_dict_org_1) #Status dict has all visited nodes
    org_1_stuck_internal_single = []
    org_1_stuck_external = []
    for nodes in org_1_stuck_nodes_internal:
        if 'IR' in nodes or 'RR' in nodes or 'RevBR' in nodes:
            if 'E' not in nodes:
                org_1_stuck_internal_single.append(nodes)
            else:
                org_1_stuck_external.append(nodes)
        else:
            org_1_stuck_external.append(nodes)
    print('stuck_external_rxns_in_org_1_combined ' + org_1, len(stuck_external_rxns_in_org_1_combined))
    print('stuck_external_rxns_in_org_2_combined ' + org_2, len(stuck_external_rxns_in_org_2_combined))
    print('stuck_internal_rxns_in_org_2_combined ' + org_2, len(stuck_internal_rxns_in_org_2_combined))
    print('stuck_internal_rxns_in_org_1_combined ' + org_1, len(stuck_internal_rxns_in_org_1_combined))
    print('org_1_stuck_internal_single ' + org_1, len(org_1_stuck_internal_single))
    print('org_2_stuck_internal_single ' + org_2, len(org_2_stuck_internal_single))
    print('org_2_stuck_external ' + org_2, len(org_2_stuck_external))
    print('org_1_stuck_external ' + org_1, len(org_1_stuck_external))
    print('Internal reactions org_1 '+ org_1, len(org_1_internal_reactions_in_combined))
    print('Internal reactions org_2 '+ org_2, len(org_2_internal_reactions_in_combined))
    print('External reactions org_1 ' + org_1, len(dummy_1))
    print('External reactions org_2 '+ org_2, len(dummy_2))
    print('Metabolic support index ' + org_1, (len(org_1_stuck_internal_single)-len(stuck_internal_rxns_in_org_1_combined))/ len(org_1_stuck_internal_single))
    print('Metabolic support index ' + org_2, (len(org_2_stuck_internal_single)-len(stuck_internal_rxns_in_org_2_combined))/ len(org_2_stuck_internal_single))
    scope_without_seed = scope - seed_metabolites
    org_1_scope = set([])
    org_2_scope = set([])
    external = set([])
    for mets in scope_without_seed:
        if org_1 in mets:
            org_1_scope.add(mets)
        elif org_2 in mets:
            org_2_scope.add(mets)
        else:
            external.add(mets)
    org_1_individual= set([])
    org1_external = set([])
    org_1_individual_scope = scope_org_1 - seed_metabolites
    for mets in org_1_individual_scope:
        if org_1 in mets:
            org_1_individual.add(mets)
        else:
            org1_external.add(mets)
    org_2_individual= set([])
    org2_external = set([])
    org_2_individual_scope = scope_org_2 - seed_metabolites
    for mets in org_2_individual_scope:
        if org_2 in mets:
            org_2_individual.add(mets)
        else:
            org2_external.add(mets)
    print('Scope ' + org_1, len(org_1_scope))
    print('Scope ' + org_2, len(org_2_scope))
    print('External', len(external))
    print('Individual scope ' + org_1, len(org_1_individual))
    print('Individual scope ' + org_2, len(org_2_individual))


if __name__ == '__main__':
   # pname = input('Enter folder name with all files \n')
   # graph_name_combined = input('Enter the name of the joint graph generated by MetQuest with extension .gpickle extension \n')
   # graph_name_firstorg = input('Enter the name of the first organism graph generated by MetQuest with extension .gpickle extension \n')
   # graph_name_secondorg = input('Enter the name of the second organism graph generated by MetQuest with extension .gpickle extension \n')
   # seed_metabolite_fname = input('Enter the name of the name of the seed metabolite filename with .txt extension \n')
   # find_metabolic_support_index(pname, graph_name_combined, graph_name_firstorg, graph_name_secondorg, seed_metabolite_fname)

    pname = path + "/"
    graph_name_combined = "Rhizobium_CF142_pacbio_gbk_genome_proteins_newids_Sphingobium_AP49_pacbio_gbk_genome_proteins_newids_Pseudomonas_GM17_pacbio_gbk_genome_proteins_newids_Variovorax_CF313_pacbio_gbk_genome_proteins_newids_.gpickle"
    graph_name_firstorg = "Pseudomonas_GM17_pacbio_gbk_genome_proteins_newids_.gpickle"
    graph_name_secondorg = "Rhizobium_CF142_pacbio_gbk_genome_proteins_newids_.gpickle"
    seed_metabolite_fname = "seeds_met.txt"

    msi = find_metabolic_support_index(pname, graph_name_combined, graph_name_firstorg, graph_name_secondorg, seed_metabolite_fname)

```
